# Panorama do COVID-19 no Brasil
COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

Vale ressaltar, que analisar uma doença nova pode ser extremamente desafiador, pois muitas vezes há pouca informação disponível sobre a doença em questão, incluindo como ela se espalha, quais são seus sintomas e como tratá-la.

Quando uma nova doença surge, é comum haver conflitos de dados entre governos federal, estaduais e fontes não oficiais. Os governos podem ter diferentes metodologias para coletar e relatar dados sobre a doença, o que pode levar a divergências nos números e na interpretação dos dados. Isso pode gerar confusão e desconfiança nos dados divulgados, além de prejudicar a tomada de decisões baseadas em dados.

Além disso, fontes não oficiais podem relatar dados diferentes dos governos, o que pode gerar ainda mais conflito. Algumas fontes não oficiais podem não ter acesso aos mesmos recursos ou informações que os governos, o que pode afetar a precisão dos dados que elas coletam e divulgam.

Para minimizar os conflitos de dados e a desinformação, é importante que os governos se comuniquem claramente sobre como estão coletando e relatando dados, e que forneçam informações precisas e atualizadas ao público. Também é importante incentivar a transparência nos dados e encorajar a colaboração entre governos, organizações e fontes não oficiais.

## Dicionário de variáveis:


* `iso_code` - Os países são nomeados por códigos de 3 letras com base na norma ISO 3166-1 alfa-3.
* `continent` - Continente.
* `location` - País.
* `date` - Data de observação.
* `total_cases` - Total de casos confirmados de COVID-19.
* `new_cases` - Novos casos confirmados de COVID-19. Em casos raros em que a fonte dos dados relata uma alteração diária negativa devido a uma correção de dados, o valor é difinido como NAN.
* `new_cases_smoothed` - Novos casos confirmados de COVID-19 (suavização de 7 dias).
* `total_deaths` - Total de mortes atribuídas à COVID-19.
* `new_deaths` - Novas mortes atribuídas à COVID-19. Em casos raros em que a fonte dos dados relata uma alteração diária negativa devido a uma correção de dados, o valor é difinido como NAN.
* `new_deaths_smoothed` -  Novas mortes atribuídas ao COVID-19 (suavização de 7 dias).
* `total_cases_per_million` - Total de casos confirmados de COVID-19 por 1.000.000 de pessoas.
* `new_cases_per_million` - Novos casos confirmados de COVID-19 por 1.000.000 de pessoas.
* `new_cases_smoothed_per_million` - Novos casos confirmados de COVID-19 (suavização de 7 dias) por 1.000.000 de pessoas.
* `total_deaths_per_million` - Total de mortes atribuídas ao COVID-19 por 1.000.000 de pessoas.
* `new_deaths_per_million` - Novas mortes atribuídas ao COVID-19 por 1.000.000 de pessoas.
* `new_deaths_smoothed_per_million` - Novas mortes atribuídas ao COVID-19 (suavização de 7 dias) por 1.000.000 de pessoas.
* `reproduction_rate` - Estimativa em tempo real da taxa efetiva de reprodução (R) do COVID-19.
* `icu_patients` - Número de pacientes com COVID-19 em unidades de terapia intensiva (UTIs) em um determinado dia.
* `icu_patients_per_million` -  Número de pacientes com COVID-19 em unidades de terapia intensiva (UTIs) em um determinado dia por 1.000.000 de pessoas.
* `hosp_patients` - Número de pacientes com COVID-19 internados em um determinado dia
* `hosp_patients_per_million` - Número de pacientes com COVID-19 hospitalizados em um determinado dia por 1.000.000 de pessoas
* `weekly_icu_admissions` - Número de pacientes com COVID-19 recém-admitidos em unidades de terapia intensiva (UTIs) em uma determinada semana (data do relatório e os 6 dias anteriores). 
* `weekly_icu_admissions_per_million` - Número de pacientes com COVID-19 recém-admitidos em unidades de terapia intensiva (UTIs) em uma determinada semana por 1.000.000 de pessoas (data do relatório e os 6 dias anteriores).
* `weekly_hosp_admissions` - Número de pacientes com COVID-19 recém-admitidos em hospitais em uma determinada semana (data do relatório e os 6 dias anteriores).
* `weekly_hosp_admissions_per_million` - Número de pacientes com COVID-19 recém-admitidos em hospitais em uma determinada semana por 1.000.000 de pessoas (data do relatório e os 6 dias anteriores)
* `total_tests` - Total de testes para COVID-19.
* `new_tests` - Novos testes para COVID-19 (calculados apenas para dias consecutivos).
* `total_tests_per_thousand` - Total de testes para COVID-19 por 1.000 pessoas.
* `new_tests_per_thousand` - Novos testes para COVID-19 por 1.000 pessoas.
* `new_tests_smoothed` - Novos testes para COVID-19 (7 dias suavizados).
* `new_tests_smoothed_per_thousand` - Novos testes para COVID-19 (7 dias suavizados) por 1.000 pessoas
* `positive_rate` - A parcela de testes positivos para COVID-19.
* `tests_per_case` - estes realizados por novo caso confirmado de COVID-19.
* `tests_units` -  Unidades usadas pelo local para relatar seus dados de teste. Um arquivo de país não pode conter unidades mistas. Todas as métricas relativas aos dados de teste usam a unidade de teste especificada. As unidades válidas são 'pessoas testadas' (número de pessoas testadas), 'testes realizados' (número de testes realizados. uma única pessoa pode ser testada mais de uma vez em um determinado dia) e 'amostras testadas' (número de amostras testadas. Em alguns casos, mais de uma amostra pode ser necessária para realizar um determinado teste.).
* `total_vaccinations` - Número total de doses de vacinação contra COVID-19 administradas.
* `people_vaccinated` - Número total de pessoas que receberam pelo menos uma dose de vacina.
* `people_fully_vaccinated` - Número total de pessoas que receberam todas as doses prescritas pelo protocolo inicial de vacinação.
* `total_boosters` - Número total de doses de reforço da vacinação contra COVID-19 administradas.
* `new_vaccinations` - Novas doses de vacinação COVID-19 administradas.
* `new_vaccinations_smoothed` - Novas doses de vacinação COVID-19 administradas (suavização de 7 dias).
* `total_vaccinations_per_hundred` - Número total de doses de vacinação contra COVID-19 administradas por 100 pessoas na população total.
* `people_vaccinated_per_hundred` - Número total de pessoas que receberam pelo menos uma dose de vacina por 100 pessoas na população total.
* `people_fully_vaccinated_per_hundred` - Número total de pessoas que receberam todas as doses prescritas pelo protocolo inicial de vacinação por 100 pessoas na população total.
* `total_boosters_per_hundred` - Número total de doses de reforço da vacinação contra COVID-19 administradas por 100 pessoas na população total.
* `new_vaccinations_smoothed_per_million` -  Novas doses de vacinação contra COVID-19 administradas (suavizadas em 7 dias) por 1.000.000 de pessoas na população total.
* `new_people_vaccinated_smoothed` -  Número diário de pessoas que receberam a primeira dose da vacina (7 dias suavizados).
* `new_people_vaccinated_smoothed_per_hundred` - Número diário de pessoas que receberam a primeira dose da vacina (suavização de 7 dias) por 100 pessoas na população total.
* `stringency_index` - Ìndice de Severidade de Resposta do Governo: medida composta com base em 9 indicadores de resposta, incluindo fechamento de escolas, fechamento de locais de trabalho e proibições de viagens. valor de 0 a 100 (100 = resposta mais rigorosa).
* `population` - População (últimos valores disponíveis).
* `population_density` - Número de pessoas dividido por área de terra, medida em quilômetros quadrados, ano mais recente disponível.
* `median_age` - Idade média da população, projeção da ONU para 2020.
* `aged_65_older` - Parcela da população com 65 anos ou mais, ano mais recente disponível
* `aged_70_older` - Proporção da população com 70 anos ou mais em 2015.
* `gdp_per_capita` - Produto interno bruto em paridade de poder de compra (dólares internacionais constantes de 2011), ano mais recente disponível.
* `extreme_poverty` - Parcela da população que vive em extrema pobreza, ano mais recente disponível desde 2010.
* `cardiovasc_death_rate` - Taxa de mortalidade por doenças cardiovasculares em 2017 (número anual de mortes por 100.000 pessoas).
* `diabetes_prevalence` - Prevalência de diabetes (% da população de 20 a 79 anos) em 2017 .
* `female_smokers` - Proporção de mulheres que fumam, ano mais recente disponível.
* `male_smokers` - Parcela de homens que fumam, ano mais recente disponível.
* `handwashing_facilities` -  Parcela da população com instalações básicas de lavagem das mãos nas instalações, ano mais recente disponível.
* `hospital_beds_per_thousand` - Camas hospitalares por 1.000 pessoas, ano mais recente disponível desde 2010
* `life_expectancy` - Expectativa de vida ao nascer em 2019
* `human_development_index` - Um índice composto que mede o desempenho médio em três dimensões básicas do desenvolvimento humano – uma vida longa e saudável, conhecimento e um padrão de vida decente. Valores para 2019.
* `excess_mortality` - Diferença percentual entre o número relatado de mortes semanais ou mensais em 2020–2021 e o número projetado de mortes para o mesmo período com base nos anos anteriores.
* `excess_mortality_cumulative` - Diferença percentual entre o número acumulado de mortes desde 1º de janeiro de 2020 e as mortes projetadas cumulativas para o mesmo período com base nos anos anteriores.
* `excess_mortality_cumulative_absolute` -  Diferença cumulativa entre o número relatado de mortes desde 1º de janeiro de 2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores. 
* `excess_mortality_cumulative_per_million` -  Diferença cumulativa entre o número relatado de mortes desde 1º de janeiro de 2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores, por milhão de pessoas.

## Analise dos dados 
*Em seguida, vamos analisar os os dados e ver os insight gerado. Inicialmente vamos analisar os países com maior e menor número de mortes que ocorreram até o final de 2020. Já que foi o ano em que o covid atingiu grande parte do mundo.*

```python
# Inicialmente vamos obversar os paises com maiores números de morte
df.loc[df.date == '2020-12-31'].sort_values(by="total_deaths", ascending=False)
```

```python
# Em seguida, como nosso objetivo é analisar os países, vamos criar 2  dataset para selecionar os 7 países com maior número de mortes da coluna "location".
df_1 = df.loc[df.date == '2020-12-31'].sort_values(by="total_deaths", ascending=False)[10:16]
df_2 = df.loc[df.total_deaths == 350568.0].sort_values(by="total_deaths", ascending=False)

# Agora vamos juntas os dataframes criados 
top_7_total_deaths = pd.merge(df_2, df_1, how = 'outer')

# exibir o dataframe
top_7_total_deaths[['location', 'total_deaths']]
```

```python
# Para melhorar a visualização vamos criar um gráfico
fig, ax = plt.subplots(figsize = (10,5))
ax.bar('location',"total_deaths", data=top_7_total_deaths)
ax.set_xlabel('Países')
ax.set_ylabel("Número de mortes")
ax.set_title("Número de mortes por países")

fig.tight_layout()
```

*Portanto, podemos notar que o Estados Unidos, possui o maior número de mortes, quase o dobro do Brasil que está em segundo lugar.*

### Como o vírus está progredindo?
A evolução dos casos e mortes relacionados à Covid-19 tem sido uma das principais preocupações das organizações de saúde em todo o mundo. Para visualizar essa evolução ao longo do tempo, uma das melhores formas é através do gráfico de linhas.

Nesse sentido, podemos criar um gráfico de linhas para demonstrar a evolução do número de mortes em todo o mundo. Essa ferramenta permite uma análise visual mais clara e objetiva, facilitando a identificação de tendências e mudanças ao longo do tempo.

Com isso, podemos monitorar a evolução da pandemia e compreender melhor o impacto das medidas tomadas para controlar a disseminação do vírus

```python
# Plotarei um gráfico de linhas para o número de mortes no mundo 
fig, ax = plt.subplots(figsize = (11,6))
df.loc[df.location == 'World', 'total_deaths'].reset_index(drop=True).plot(kind='line', ax=ax, linestyle = 'dotted', color = 'r', linewidth = '2')
ax.set_ylabel('Número de mortes')

fig.tight_layout()
```

#Análise Exploratória para o Brasil

Agora irei realizar uma analise dos dado, direcionado ao Brasil. Inicialmente irei criar uma dataframe que contenha apenas os dados do Brasil.


```python
# Criação do novo dataframe
df_brazil = df.loc[df.location == 'Brazil'].copy()
df_brazil.head()
```

Agora irei analisar o primeiro caso de covid-19 e a primeira morte.

```python
 # Identificando a primeira morte
 df_brazil.loc[df_brazil.total_deaths == 1]
```

```python
# datas da primeira morte e caso
print(f'O primeiro caso de covid ocorreu em: {df_brazil.loc[31483].date}')
print(f'E a primeira morte ocorreu em: {df_brazil.loc[31503].date}\n')

# Quantos dias demorou para ocorrencia da primeira morte, após o primeiro caso
from datetime import date

time_dif = df_brazil.loc[31503].date - df_brazil.loc[31483].date
print(f'Demorou {time_dif.days} dias para a primeira morte, após a detectação do primeiro caso de Covid-19')
```

Agora irei plotar alguns gráficas para melhorar o entendimento de forma visual sobre o avanço do número de casos e mortes no Brasil

```python
# Primeiro irei observar quantos dias corridos tem os dados do dataframe
df_dias = df_brazil.date.max() - df_brazil.loc[31483].date

print((f'O dataframe possui dados que representam {df_dias.days} dias'))
```

```python
# gráfico de linha para número de mortes em escala linear e logarítmica
b = (df_dias.days) + 1
x = list(range(b))
y1 = df_brazil['total_cases']
y2 = df_brazil['total_deaths']

fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows = 2, ncols= 2, figsize = (15,10))

# Total de casos
ax1.plot(x,y1)
ax1.set_title('Casos no Brasil')
ax1.set_ylabel('Número de casos')
ax1.set_xlabel('Dias')

ax2.plot(x,y1)
ax2.set_title('Casos no Brasil (Escala Logarítmica)')
ax2.set_ylabel('Número de casos')
ax2.set_xlabel('Dias')
ax2.set_yscale('log')

#total de mortes
ax3.plot(x,y2)
ax3.set_title('Mortes no Brasil')
ax3.set_ylabel('Número de mortes')
ax3.set_xlabel('Dias')

ax4.plot(x,y2)
ax4.set_title('Mortes no Brasil (Escala Logarítmica)')
ax4.set_ylabel('Número de mortes')
ax4.set_xlabel('Dias')
ax4.set_yscale('log')

plt.tight_layout()
```

Agora irei analisar a relação entra o total da população e quanto dessa população tomaram alguma dose da vacina


```python
# Irei criar uma dataframe que contenham as quantidades maximas para cada coluna citada no codigo abaixo
df_vacina = df_brazil[['population', 'people_vaccinated', 'people_fully_vaccinated' ,'total_boosters' ]].sort_values(by='total_boosters', ascending = False)[:1]
# mostra o resultado do dataframe criado
df_vacina
```

```python

# Criar uma figura e um objeto de eixos
fig, ax = plt.subplots(figsize = (15,10))

# Criar as barras usando o método bar do objeto de eixos
ax.bar(df_vacina.columns, df_vacina.values[0])

# Adicionar títulos e rótulos
ax.set_title('Dados de Vacinação')
ax.set_xlabel('População e quantidade de doses')
ax.set_ylabel('Quantidade')

# Exibir o gráfico
plt.show()
```

Uma pequena legenda:


*   people_vaccinated - Número total de pessoas que receberam pelo menos uma dose de vacina.
*   people_fully_vaccinated - Número total de pessoas que receberam todas as doses prescritas pelo protocolo inicial de vacinação.
*   total_boosters - Número total de doses de reforço da vacinação contra COVID-19 administradas.
*   population - População do país

Portanto, quando analisamos o gráfico a cima percebemos que nem toda parcela da população brasileira tomou a primeira dose da vacina contra o Covid-19 e que para as doses seguintes (segunda dose e reforço) o número de pessoas a tomar as doses foram diminuindo.
